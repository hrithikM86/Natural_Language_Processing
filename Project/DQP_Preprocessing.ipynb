```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import re  #regular expressions
```

```python
df=pd.read_csv('train.csv')
```

```python
#Lowercasing
```

```python
df['question1']=df['question1'].str.lower()
```

```python
df['question2']=df['question2'].str.lower()
```

```python
df.head()
```

```python
#Removing HTML tags
```

```python
def remove_html_tags(text):
    pattern = re.compile('<.*?>')              #regular expression to find HTML tags = <.*?>
    return pattern.sub(r'', str(text))
```

```python
df['question1']=df['question1'].apply(remove_html_tags)
```

```python
df['question2']=df['question2'].apply(remove_html_tags)
```

```python
#Removing URLS
```

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', str(text))
```

```python
df['question1']=df['question1'].apply(remove_url)
```

```python
df['question2']=df['question2'].apply(remove_url)
```

```python
#Removing punctuations
```

```python
import string
```

```python
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
        
```

```python
df['question1']=df['question1'].apply(remove_punc)
```

```python
df['question2']=df['question2'].apply(remove_punc)
```

```python
def remove_stopwords(text):
    new_text = []
    
    for word in text.split():
        if word in sw_:
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```
